## 8. Escribe una funci贸n que devuelva la moda de una lista, es decir: el elemento que aparece m谩s veces.

`NOTE: No se debe usar count... `

```python
def mode_counter(arr):
    dict_cnt = dict()
    mode = 0

    for key in arr:
        if key in dict_cnt:
            dict_cnt[key] += 1
        else:
            dict_cnt[key] = 1

        if dict_cnt[key] > mode:
            mode = key
                
    return mode
```

```python
mode_counter([1,2,3,3,4,5,6])
```

```python
# Para testear tu funci贸n 
test_mode(mode_counter)
```

La f贸rmula es esta para la Sample Standard Deviation:

Si quiesieramos la Population, cambiar铆amos (N-1) por N

```python
def st_dev(list_sd):
    N = my_len(list_sd)

    mean = sum_all(list_sd) / my_len(list_sd) 
    variance = sum_all([((xi - mean)**2) for xi in list_sd]) / (N-1)
    
    return variance**0.5
```

```python
import re
def sort_alpha(str_commas):
    pattern = re.compile(r'[^,]+')
    words = sorted(pattern.findall(str_commas))
    sorted_words = sorted(words)

    sort_str_commas = ""
    for idx, word in enumerate(sorted_words):
        sort_str_commas += word
        if idx != (my_len(sorted_words)-1):
            sort_str_commas += ","        

    return sort_str_commas
```
