# Introducci√≥n a ML

## ¬øQu√© es el Machine Learning?

+ El aprendizaje autom√°tico (ML) consiste en crear programas que aprenden de los datos para mejorar su rendimiento autom√°ticamente.

+ **Hay tres tipos principales:**
    + **Supervisado** (aprenden con datos etiquetados)
    + **No supervisado** (agrupan datos similares sin etiquetas)
    + **Por refuerzo** (mejoran probando y recibiendo retroalimentaci√≥n). Nos centraremos en los dos primeros.

### **Apredizaje Supervisado**

+ En este caso, el algoritmo aprende a partir de un conjunto de datos en los que las entradas (datos de entrada) est√°n emparejadas con sus salidas esperadas (etiquetas). El objetivo es aprender la relaci√≥n para predecir etiquetas en datos nuevos.

**Ejemplo: Clasificaci√≥n de correos como spam o no spam**

+ **Entrada**: Una base de datos de correos electr√≥nicos con informaci√≥n como palabras frecuentes, enlaces, y caracter√≠sticas adicionales.
+ **Etiqueta**: Cada correo est√° etiquetado como "Spam" o "No Spam".
+ **Uso**: El modelo aprende a identificar patrones asociados con los correos spam y puede clasificar nuevos correos en una de estas dos categor√≠as.

**Mas Ejemplos**

- Predicci√≥n de precios de casas: Usar caracter√≠sticas como tama√±o, ubicaci√≥n, y n√∫mero de habitaciones para predecir el precio.
- Reconocimiento de im√°genes: Clasificar im√°genes como "gato" o "perro" bas√°ndose en datos etiquetados.
- Diagn√≥stico m√©dico: Identificar enfermedades como "diabetes" o "hipertensi√≥n" a partir de datos m√©dicos etiquetados.


### **Aprendizaje No Supervisado**

+ Aqu√≠, el algoritmo trabaja con datos que no tienen etiquetas. El objetivo es descubrir patrones o estructuras ocultas, como grupos (cl√∫steres) o distribuciones.

**Ejemplo: Segmentaci√≥n de clientes en un supermercado**

+ **Entrada**: Datos sobre las compras de los clientes, como frecuencia de visitas, monto promedio gastado, y productos comprados.
+ **Etiqueta**: No hay etiquetas (no sabemos a qu√© grupo pertenece cada cliente de antemano).
+ **Uso**: El modelo agrupa autom√°ticamente a los clientes en cl√∫steres, como "compradores frecuentes", "compradores de lujo", o "compradores de ofertas", lo que ayuda a personalizar promociones o servicios.

**Mas Ejemplos**

- Agrupaci√≥n de genes: Encontrar patrones en datos gen√©ticos para identificar grupos de genes con funciones similares.
- Recomendaciones de pel√≠culas: Agrupar usuarios seg√∫n preferencias sin etiquetas expl√≠citas para recomendar contenido.
- An√°lisis de fraudes: Detectar transacciones inusuales agrupando patrones en los datos sin conocer previamente qu√© es fraudulento.


### **Aprendizaje por Refuerzo**

+ El algoritmo aprende interactuando con un entorno. Realiza acciones, recibe retroalimentaci√≥n en forma de recompensas o penalizaciones, y ajusta su estrategia para maximizar la recompensa acumulada.

**Ejemplo: Un robot aprendiendo a caminar**

+ **Entrada**: Sensores del robot que le indican su estado actual (posici√≥n, velocidad, equilibrio, etc.).
+ **Acci√≥n**: El robot decide mover sus piernas en diferentes direcciones.
+ **Retroalimentaci√≥n**: Si el robot se mantiene de pie o avanza correctamente, recibe una recompensa. Si cae, recibe una penalizaci√≥n.
+ **Uso**: El robot aprende progresivamente a ajustar sus movimientos para caminar de manera m√°s eficiente.

**Mas Ejemplos**

- Un coche aut√≥nomo aprendiendo a conducir: Ajusta su velocidad y direcci√≥n para maximizar seguridad y eficiencia en el tr√°fico.
- Jugadores de videojuegos: Un agente aprende a ganar puntos en un juego probando estrategias y recibiendo recompensas.
- Optimizaci√≥n de rutas de entrega: Un algoritmo aprende a planificar rutas que minimicen tiempo y costos al recibir recompensas por eficiencia.


## **Flujo T√≠pico de un Proyecto de Aprendizaje Autom√°tico**

Cuando desarrollamos un proyecto de aprendizaje autom√°tico, seguimos varias etapas clave. Aqu√≠ tienes una breve descripci√≥n de cada paso:

1. üìÇ **Datos crudos** ‚Üí üõ†Ô∏è **Preprocesamiento** ‚Üí ‚úÇÔ∏è **Divisi√≥n** ‚Üí ü§ñ **Entrenamiento** ‚Üí üìä **Evaluaci√≥n** ‚Üí üìà **Predicci√≥n**

----

### 1. **Obtenci√≥n de Datos**
   - Lo primero es contar con un conjunto de datos representativo. Esto puede provenir de:
     - Archivos CSV o bases de datos.
     - Fuentes p√∫blicas como datasets en Kaggle o UCI Machine Learning Repository.
     - Datos generados por sensores o aplicaciones.

### 2. **Preprocesamiento**
   - Los datos reales suelen estar desordenados, incompletos o con valores no v√°lidos.  
     **Ejemplo de tareas comunes de preprocesamiento:**
     - Manejar valores nulos o duplicados.
     - Escalar caracter√≠sticas (normalizaci√≥n o estandarizaci√≥n).
     - Codificar variables categ√≥ricas en formato num√©rico.

### 3. **Divisi√≥n de Datos**
   - Para entrenar y evaluar correctamente un modelo, dividimos el conjunto de datos en:
     - **Entrenamiento:** Datos para que el modelo aprenda (70%-80%).
     - **Prueba:** Datos para evaluar la precisi√≥n del modelo (20%-30%).

### 4. **Entrenamiento del Modelo**
   - Usamos un algoritmo adecuado para el tipo de problema:
     - **Clasificaci√≥n:** Identificar categor√≠as (ejemplo: spam/no spam).
     - **Regresi√≥n:** Predecir valores continuos (ejemplo: precios de casas).
     - **Clustering:** Agrupar datos similares (ejemplo: segmentaci√≥n de clientes).

### 5. **Evaluaci√≥n del Modelo**
   - Comparamos las predicciones del modelo con los valores reales usando m√©tricas como:
     - **Clasificaci√≥n:** Precisi√≥n (accuracy), matriz de confusi√≥n, F1-score.
     - **Regresi√≥n:** Error Absoluto Medio (MAE), Error Cuadr√°tico Medio (MSE).

### 6. **Predicci√≥n con Datos Nuevos**
   - Finalmente, usamos el modelo entrenado para hacer predicciones sobre datos que no ha visto antes.

2. **Coeficientes:** Cada coeficiente muestra c√≥mo cambia la **progresi√≥n de la diabetes** (variable de salida) cuando una caracter√≠stica cambia en **una unidad**, manteniendo las dem√°s constantes.

 - **Coeficiente de la edad (x1 = -10.01):**  
   Por cada **a√±o adicional de edad**, la progresi√≥n de la diabetes disminuye en **10.01 unidades**. Es decir, a medida que la edad aumenta, la progresi√≥n de la diabetes **disminuye ligeramente** seg√∫n este modelo.
 
 - **Coeficiente del √≠ndice de masa corporal (x2 = -239.82):**  
   Por cada **aumento de 1 unidad en el √≠ndice de masa corporal (IMC)**, la progresi√≥n de la diabetes **disminuye en 239.82 unidades**. Este valor negativo puede parecer contraintuitivo, pero podr√≠a ser el efecto de la relaci√≥n entre el IMC y otras variables del modelo.

   Por cada **aumento de 1 unidad en la presi√≥n arterial media**, la progresi√≥n de la diabetes **aumenta en 519.84 unidades**. Esto sugiere que **una mayor presi√≥n arterial** est√° asociada con un **aumento en la progresi√≥n de la diabetes**.

 - Los dem√°s coeficientes se interpretan de forma similar: indican el cambio en la progresi√≥n de la diabetes por cada unidad adicional en esas caracter√≠sticas espec√≠ficas, ya sea **positiva** o **negativa**.

#### **En resumen:**
- **Intercepto:** 152.13 es el valor base de la progresi√≥n de la diabetes cuando todas las caracter√≠sticas son cero.
- **Coeficientes:** Indican cu√°nto cambia la progresi√≥n de la diabetes por cada **unidad** de cambio en las caracter√≠sticas (edad, √≠ndice de masa corporal, etc.). Los valores positivos indican un aumento en la progresi√≥n, y los valores negativos indican una disminuci√≥n.

Una vez que hemos generado un modelo lineal, podemos probar cada coeficiente utilizando una prueba t para ver si el intervalo de confianza de la variable contiene cero. Tambi√©n podemos realizar una prueba F global para comprobar si al menos un coeficiente es significativamente distinto de cero.

## 1. Importar las bibliotecas necesarias:

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

## 2. Preparar los datos para statsmodels:

El paquete statsmodels requiere que agreguemos una columna de 1s a los datos para representar el t√©rmino de intercepto (similar a lo que hace sklearn, pero expl√≠citamente aqu√≠).

```python
# Agregar una columna de 1s para el intercepto
X = sm.add_constant(diabetes.data)
y = diabetes.target
```

## 3. Ajustar el modelo con statsmodels:

Ahora que tenemos los datos listos, ajustamos el modelo usando OLS (m√≠nimos cuadrados ordinarios), que es el m√©todo utilizado por statsmodels para la regresi√≥n lineal.


```python
# Ajustar el modelo OLS (m√≠nimos cuadrados ordinarios)
model = sm.OLS(y, X)
results = model.fit()
```

## 4. Imprimir el resumen de los resultados:

Para obtener un resumen completo de los resultados, utilizamos el m√©todo .summary(). Este resumen incluir√° los coeficientes, sus errores est√°ndar, los valores p (que utilizaremos para las pruebas t) y otras m√©tricas importantes.

```python
# Imprimir el resumen de los resultados
print(results.summary())
```

## 5. Interpretar los resultados

El resumen incluir√° una tabla con columnas como:

+ **Coef.: los coeficientes del modelo.** Los coeficientes del modelo indican cu√°nto cambia la variable objetivo cuando la caracter√≠stica correspondiente aumenta en una unidad, manteniendo las dem√°s constantes.
+ **Std Err.: el error est√°ndar de cada coeficiente.** El error est√°ndar mide la variaci√≥n de cada coeficiente; valores m√°s bajos indican estimaciones m√°s precisas.
+ **t: el valor de la estad√≠stica t para la prueba de hip√≥tesis.** El valor de la estad√≠stica t se utiliza para evaluar si el coeficiente es significativamente distinto de cero; se calcula como 

+ **P>|t|:** el valor p asociado con cada coeficiente. 


Si el valor p de un coeficiente es menor que un umbral de significancia com√∫n (por ejemplo, 0.05), entonces rechazamos la hip√≥tesis nula de que el coeficiente es igual a cero, lo que sugiere que esa caracter√≠stica tiene una relaci√≥n significativa con la variable de respuesta.

## Que es R2?

La puntuaci√≥n R¬≤ nos indica cu√°nta variaci√≥n de la variable de salida (en este caso mpg) est√° siendo explicada por el modelo a partir de las caracter√≠sticas de entrada. **Un R¬≤ cercano a 1 indica que el modelo est√° prediciendo bien**, mientras que un valor cercano a 0 indica que el modelo no est√° haciendo predicciones mucho mejores que simplemente predecir la media de mpg.

En este caso:

 - Si la puntuaci√≥n R¬≤ es 0.8, significa que el modelo puede explicar el 80% de la variabilidad en el rendimiento de combustible utilizando las caracter√≠sticas del veh√≠culo (como el peso, la potencia del motor, etc.).
